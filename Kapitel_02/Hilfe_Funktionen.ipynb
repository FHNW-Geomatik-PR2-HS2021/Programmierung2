### Was ist ein `docstring`?

```python
def func(a, b, msg=""):
    """Gibt den shape von `a` und `b` mit einem optionalen Message `msg` aus.
    
    Dies ist eine recht nutzlose Funktion, die demonstriert, wie ein docstring
    geschrieben wird (dieser folgt der numpy-docstring-style Konvetion:
    https://numpydoc.readthedocs.io/en/latest/format.html)
    
    Parameters
    ----------
    a : array_like
        Beiebiges N-D Array mit beliebige Dimension
    b : array_like
        Beiebiges N-D Array mit beliebige Dimension 

    msg : str, optional
        Ein Message fÃ¼r die Ausgabe. Standardwert ist eine Leere Zeichenkette.
        
        
    Beispiel
    --------
    >>> a = np.zeros(shape=(5,5,5))
    >>> b = np.zeros(shape=(1,3,3,10))
    >>> func(a, b, msg="Ergebnis:")a = np.zeros()
    Ergebnis: (5, 5, 5) (1, 3, 3, 10)
    """
    
    # dies ist der eigenliche Code
    print(msg, a.shape, b.shape)
```

```python
a = np.zeros(shape=(5,5,5))
b = np.zeros(shape=(1,3,3,10))

func(a, b, msg="Ergebnis:")
```
